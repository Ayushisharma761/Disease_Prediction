```python
# Drop the 'sno' column from the DataFrame
#import_sheet_df.drop('slno', axis=1, inplace=True)

# List of column names to drop
columns_to_drop = ['slno' ,  'Month']

# Drop the specified columns from the DataFrame
import_sheet_df = import_sheet_df.drop(columns_to_drop, axis=1)
```

```python
dup_cols= set()

for x in range (import_sheet_df.shape[1]):
    base_col = import_sheet_df.iloc[:,x]
    for y in range(x+1, import_sheet_df.shape[1]):
        comp_col = import_sheet_df.iloc[:,y]
        
        if base_col.equals(comp_col):
            dup_cols.add(import_sheet_df.columns.values[y])



print(dup_cols)
```

```python
import_sheet_df.drop(labels=dup_cols,axis=1)
```
